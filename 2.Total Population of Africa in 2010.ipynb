```python
# Calculate the total population for all African countries in 2010.
# Create a bar chart showing the population distribution across African countries in 2010


import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('population.csv')

# Filter the data for the year 2010 and continent Africa
data_2010_africa = data[(data['year'] == 2010) & (data['continent'] == 'Africa')]

# Calculate the total population for all African countries in 2010
total_population_africa_2010 = data_2010_africa['population'].sum()
print(f"Total population of African countries in 2010: {total_population_africa_2010} million")

# Create a bar chart showing population distribution across African countries in 2010
plt.figure(figsize=(12, 6))
plt.bar(data_2010_africa['country name'], data_2010_africa['population'], color='skyblue')
plt.xticks(rotation=90, fontsize=8)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Population (in millions)', fontsize=12)
plt.title('Population Distribution Across African Countries in 2010', fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()


# Summary
# In 2010, the total population across all African countries was approximately 991 million. 
# This represents the cumulative population of the continent, highlighting Africa's demographic significance in the global context.
```
